```python
import tensorflow as tf
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler 
```

```python
print(tf.__version__)
```

```python
data = pd.read_csv('drive/My Drive/Colab Notebooks/GOOG.csv', date_parser=True)
data.head()
```

```python
data.shape
```

```python
data_training = data[data['Date']<'2019-01-01'].copy()
data_training
```

```python
data_test = data[data['Date']>='2019-01-01'].copy()
data_test
```

```python
training_data = data_training.drop(['Date', 'Adj Close'], axis=1)
training_data.head()
```

```python
scaler = MinMaxScaler()
training_data = scaler.fit_transform(training_data)
training_data
```

```python
X_train = []
y_train = []
```

```python
training_data.shape
```

```python
for i in range(60, training_data.shape[0]):
    X_train.append(training_data[i-60:i])
    y_train.append(training_data[i, 0])
```

```python
X_train, y_train = np.array(X_train), np.array(y_train)
```

```python
X_train.shape, y_train.shape
```

```python
from tensorflow.keras import Sequential 
from tensorflow.keras.layers import Dense, LSTM, Dropout 
```

```python
regressor = Sequential()

regressor.add(LSTM(units = 50, activation='relu', return_sequences=True, input_shape=(X_train.shape[1], 5)))
regressor.add(Dropout(0.2))

regressor.add(LSTM(units = 60, activation='relu', return_sequences=True))
regressor.add(Dropout(0.3))

regressor.add(LSTM(units = 80, activation='relu', return_sequences=True))
regressor.add(Dropout(0.4))

regressor.add(LSTM(units = 120, activation='relu'))
regressor.add(Dropout(0.5))

regressor.add(Dense(units = 1))
```

```python
regressor.summary()
```

```python
adam = tf.keras.optimizers.Adam(
    learning_rate=0.0003, beta_1=0.9, beta_2=0.999, epsilon=1e-07, name='Adam'
)
```

```python
regressor.compile(optimizer=adam, loss='mean_squared_error')
```

```python
history = regressor.fit(X_train, y_train, epochs=10, batch_size=128, validation_split=0.15)
```

```python
plot_df = pd.DataFrame(history.history)
plot_df.plot()
```
